```python
def calc_distance(X, Y): 
    "The distance between two points."
    return abs(sum(X - Y))

def tour_length(X_Y_df, tour_list):
    "The total of distances between each pair of consecutive cities in the tour."
    X_Y_arr = X_Y_df.values
    distance_list=[]
    for start_index in range(len(tour_list)-1):
      start_point=tour_list[start_index]
      stop_point = tour_list[start_index+1]

      distance = calc_distance(X_Y_arr[start_point], X_Y_arr[stop_point])
      distance_list.append(distance)

    return sum(distance_list), tour_list
```

```python
def shortest_tour_after_replications(n_replications, X_Y_df): 
    "Choose the tour with the minimum tour length."

    distance_tour_list = []

    for i in range(n_replications):

      tour_list = list(range(50))
      np.random.shuffle(tour_list)
      sum_distance, tours = tour_length(X_Y_df, tour_list)
      distance_tour_list.append((sum_distance,tours))

    return sorted(distance_tour_list, key=lambda x:x[0])[0]
```

```python
min_distance, tours= shortest_tour_after_replications(100000, X_Y_df)
min_distance, tours
```

```python
#sort the df in order
final_df_sort = df.iloc[tours,:]
final_df_sort['lat'] = df['lat'].astype(float)
final_df_sort['lon'] = df['lon'].astype(float)
final_df_sort.head()
```
